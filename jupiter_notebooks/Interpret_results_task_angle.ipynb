```python
from utils.model_utils import split_equal_test_size, split_equal_test_size_task
def make_train_test_split_xgboost(data, test_size, seed, task_angle=None):
    """
    Makes train and test splits for XGBoost models.

    Args:
        - data (df): dataframe containing all data
        - test_size (float): size of test split
        - seed (int): random_state for reproducinility

    Returns:
        - X_train (df): training split
        - X_test (df): test split
        - y_train (df): training labels
        - y_test (df): test labels
    """
    # shuffle data
    for col in data.columns:
        np.random.seed(seed)
        np.random.shuffle(data[col].values)

    # define X and y
    X = data.loc[:, data.columns != "binary_label"]
    y = data["binary_label"]

    if task_angle is None or task_angle == False:
        X_train, X_test, y_train, y_test = split_equal_test_size(X, y, test_size)
        
    elif task_angle == True:
        # split into train and test using split_equal_size function to make test-set equally balanced, i.e. 50/50 depressed and controls
        X_train, X_test, y_train, y_test = split_equal_test_size_task(X, y, test_size)
    return X_train, X_test, y_train, y_test
```

# Individual Error analysis function

```python
def error_analysis(text_features, X_test, y_test, model_name):
    X_test_raw = X_test
    X_test = X_test[text_features]
    # convert embeddings to arrays
    if text_features == ["word_embeddings"] or text_features == ["sentence_embeddings"]:
        X_test = np.array(list(X_test[text_features[0]]), dtype=np.float64)
    else:
        X_test = np.array(X_test)
    # load model
    model = joblib.load(f'../output/xgboost/best_models/{model_name}.pkl')
    # predict on test data
    y_test_pred = model.predict(X_test)

    preds = model.predict_proba(X_test)
    #RocCurveDisplay.from_predictions(y_test, preds[:, 1])
    plt.show()

    df = pd.DataFrame({"ID": X_test_raw["ID"], "Diagnosis": X_test_raw["Diagnosis"], "Data type": X_test_raw["data_type"], "True label": y_test, "Prediction": y_test_pred, "Proba": preds[:, 1]})

    return df
```

### Get error analysis for all seeds

```python
# Loop through each seed and make a df with the mean prediction across seed
seeds = [1, 66, 31, 18, 97]
test_size = 0.2
df_all_tri = pd.DataFrame(columns = ["ID", "Diagnosis", "Data type", "True label", "Prediction_lowlevel", "Prediction_word_embeddings", "Prediction_sentence_embeddings", "Auto_lowlevel_predictions", "Auto_word_predictions", "Auto_sentence_predictions", "seed", "index"])
df_all_auto =  pd.DataFrame(columns = ["ID", "Diagnosis", "Data type", "True label", "Prediction_lowlevel", "Prediction_word_embeddings", "Prediction_sentence_embeddings", "Triangle_lowlevel_predictions", "Triangle_word_predictions","Triangle_sentence_predictions", "seed", "index"])

for seed in seeds:
    if seed == 1:
        triangle_data, autobiographical_data, chronic_data, first_episode_data = prepare_data_xgboost_transformer(seed=5, text_features_path="../data/features/text_features.csv")
    else:
        triangle_data, autobiographical_data, chronic_data, first_episode_data = prepare_data_xgboost_transformer(seed=seed, text_features_path="../data/features/text_features.csv")
        
    # split datasets into train and test (in order to have a held-out test set)
    X_train_auto, X_test_auto, y_train_auto, y_test_auto = make_train_test_split_xgboost(data=autobiographical_data,
                                                                                        test_size=test_size, 
                                                                                        seed=seed, task_angle = True)

    X_train_tri, X_test_tri, y_train_tri, y_test_tri = make_train_test_split_xgboost(data=triangle_data,
                                                                                        test_size=test_size, 
                                                                                        seed=seed, task_angle = True)

    X_train_chronic, X_test_chronic, y_train_chronic, y_test_chronic = make_train_test_split_xgboost(data=chronic_data,
                                                                                        test_size=test_size, 
                                                                                        seed=seed)

    X_train_first_episode, X_test_first_episode, y_train_first_episode, y_test_first_episode = make_train_test_split_xgboost(data=first_episode_data,
                                                                                        test_size=test_size, 
                                                                                        seed=seed)

    text_features=["Pronouns_all_pronouns", "Past_tense_all_verbs", "Negative_all_sentiment","Positive_all_sentiment"]
    # save for df
    auto_error = error_analysis(text_features, X_test_auto, y_test_auto, f"auto_low_level_seed_{seed}")
    gen_auto_low = error_analysis(text_features, X_test_auto, y_test_auto, f"triangle_low_level_seed_{seed}")
    gen_tri_low = error_analysis(text_features, X_test_tri, y_test_tri, f"auto_low_level_seed_{seed}")

    tri_error = error_analysis(text_features, X_test_tri, y_test_tri, f"triangle_low_level_seed_{seed}")

    text_features=["word_embeddings"]
    # save for df
    auto_word = error_analysis(text_features, X_test_auto, y_test_auto, f"autobiographical_word_embeddings_seed_{seed}")
    auto_error["Prediction_word_embeddings"] = auto_word["Prediction"]
    auto_error["Proba_word_embeddings"] = auto_word["Proba"]

    gen_tri_word = error_analysis(text_features, X_test_tri, y_test_tri, f"autobiographical_word_embeddings_seed_{seed}")
    gen_auto_word = error_analysis(text_features, X_test_auto, y_test_auto, f"triangle_word_embeddings_seed_{seed}")

    tri_word = error_analysis(text_features, X_test_tri, y_test_tri, f"triangle_word_embeddings_seed_{seed}")
    tri_error["Prediction_word_embeddings"] = tri_word["Prediction"]
    tri_error["Proba_word_embeddings"] = tri_word["Proba"]

    text_features=["sentence_embeddings"]

    # save for df
    auto_sentence = error_analysis(text_features, X_test_auto, y_test_auto, f"auto_sentence_embeddings_seed_{seed}")
    auto_error["Prediction_sentence_embeddings"] = auto_sentence["Prediction"]
    auto_error["Proba_sentence_embeddings"] = auto_sentence["Proba"]

    gen_tri_sent = error_analysis(text_features, X_test_tri, y_test_tri, f"auto_sentence_embeddings_seed_{seed}")
    gen_auto_sent = error_analysis(text_features, X_test_auto, y_test_auto, f"triangle_sentence_embeddings_seed_{seed}")

    tri_sentence = error_analysis(text_features, X_test_tri, y_test_tri, f"triangle_sentence_embeddings_seed_{seed}")
    tri_error["Prediction_sentence_embeddings"] = tri_sentence["Prediction"]
    tri_error["Proba_sentence_embeddings"] = tri_sentence["Proba"]

    auto_error["Triangle_lowlevel_predictions"] = gen_auto_low["Prediction"]
    auto_error["Triangle_word_predictions"] = gen_auto_word["Prediction"]
    auto_error["Triangle_sentence_predictions"] = gen_auto_sent["Prediction"]

    auto_error=auto_error.rename(columns = {"Prediction": "Prediction_lowlevel"})
    auto_error=auto_error.rename(columns = {"Proba": "Proba_lowlevel"})
    #auto_error.to_csv(f"../output/xgboost/predictions/autobiographical_data_error_analysis_seed_{seed}.csv")

    auto_error["seed"] = [seed]*len(auto_error)
    auto_error["index"] = range(0, len(auto_error))
    df_all_auto = df_all_auto.append(auto_error, ignore_index = True)

    tri_error["Auto_lowlevel_predictions"] = gen_tri_low["Prediction"]
    tri_error["Auto_word_predictions"] = gen_tri_word["Prediction"]
    tri_error["Auto_sentence_predictions"] = gen_tri_sent["Prediction"]
    tri_error = tri_error.rename(columns = {"Prediction": "Prediction_lowlevel"})
    tri_error = tri_error.rename(columns = {"Proba": "Proba_lowlevel"})
    #tri_error.to_csv(f"../output/xgboost/predictions/triangle_data_error_analysis_seed_{seed}.csv")

    tri_error["seed"] = [seed]*len(tri_error)
    tri_error["index"] = range(0, len(tri_error))
    df_all_tri = df_all_tri.append(tri_error, ignore_index = True)
```

# Plot ROC curves

```python
from sklearn.metrics import roc_curve
tprs = []
base_fpr = np.linspace(0, 1, 101)

plt.figure(figsize=(5, 5))
plt.axes().set_aspect('equal', 'datalim')
y = df_all_tri[0:22]
y_true = y["True label"]

for i in np.unique(df_all_tri["seed"]):
    subset = df_all_tri[df_all_tri["seed"] == i]
    preds = subset["Prediction_sentence_embeddings"]

    true = np.array(y_true).astype(int)
    preds = np.array(preds).astype(int)
    

    fpr, tpr, _ = roc_curve(true, preds)

    plt.plot(fpr, tpr, 'b', alpha=0.15)
    tpr = np.interp(base_fpr, fpr, tpr)
    tpr[0] = 0.0
    tprs.append(tpr)

    
tprs = np.array(tprs)
mean_tprs = tprs.mean(axis=0)
std = tprs.std(axis=0)

tprs_upper = np.minimum(mean_tprs + std, 1)
tprs_lower = mean_tprs - std


plt.plot(base_fpr, mean_tprs, 'b')
plt.fill_between(base_fpr, tprs_lower, tprs_upper, color='grey', alpha=0.3)

plt.plot([0, 1], [0, 1],'r--')
plt.xlim([-0.01, 1.01])
plt.ylim([-0.01, 1.01])
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.show()
```
